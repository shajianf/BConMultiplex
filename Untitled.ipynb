```python
#My project on ER, tmax=1000,N=10000,ens=50,phi=uniform,mu=0.8,gama=0.2,kapa=0.5,rho0=1000 ,rho-beta(sfmyprobc.cpp)
#14.4.1401

import numpy as np
import matplotlib.pylab as plt

plt.rcParams["font.family"] ="Times New Roman"
#loading data as a matrix
ran=np.loadtxt('mypro2 sf act bet-rho Ranfee.txt')
deg = np.loadtxt('mypro2 sf act bet-rho degfee.txt')
BC = np.loadtxt('mypro2 sf act bet-rho BCfee.txt')
# each colomn as an array
fee=ran[:,0]         
rof2=ran[:,1]
rof4=deg[:,1]
rof6=BC[:,1]
#print(fee)
# Plot#ffd700,darkred,#008080,mec='orange', mec='r',mec='b'
fig, axes = plt.subplots(figsize=(12,8))

plt.plot(fee,rof2 ,  linewidth = 3,linestyle='--',marker='o',markersize=15, color='lightseagreen', label = r"$Random$")
plt.plot(fee , rof4, linewidth = 3,linestyle =':',marker='D',markersize=15 ,color="goldenrod", label = r"$Degree$")
plt.plot(fee , rof6, linewidth = 3,linestyle ='-.',marker='*',markersize=15 ,color="crimson", label = r"$Betweenness$")
plt.legend(loc="lower right", fontsize=30)
#plt.grid(color = 'C7', alpha=0.5, linestyle = ':', linewidth = 0.5)
plt.xlim([0 , 0.80])
plt.ylim([0 , 0.92])

plt.xticks([0.1 * i for i in range(9)])
plt.yticks([0.1 * i for i in range(10)])

plt.xlabel(r'$\beta^D$' ,fontsize=45)
plt.ylabel(r'$\rho^A$' , fontsize=45)

plt.xticks(fontsize=35)
plt.yticks(fontsize=35)
#plt.legend(fontsize=25)

plt.xlim(xmin=0)
plt.ylim(ymin=0)

plt.savefig('centrality act.png')
plt.savefig('centrality act.pdf')
```
